```python
#Possible draw if England finish 2nd
draw_1 = [
    [
        [
            ['France','Argentina'],
            ['Uruguay','Portugal']
        ],
        [
            ['Brazil','Mexico'],
            ['Belgium','Colombia']
        ]
    ],
    [
        [
            ['Spain','Russia'],
            ['Croatia','Denmark']
        ],
    
        [
             ['Sweden','Switzerland'],
             ['England','Japan']
        ]
    ]
]
```

```python
#Possible draw if England finish 1st
draw_2 = [
    [
        [
            ['France','Argentina'],
            ['Uruguay','Portugal']
        ],
        [
            ['Brazil','Mexico'],
            ['England','Colombia']
        ]
    ],
    [
        [
            ['Spain','Russia'],
            ['Croatia','Denmark']
        ],
    
        [
             ['Sweden','Switzerland'],
             ['Belgium','Japan']
        ]
    ]
]
```

```python
draw_4 = [
    [
        [
            ['France','Argentina'],
            ['Uruguay','Portugal']
        ],
        [
            ['Brazil','Germany'],
            ['England','Senegal']
        ]
    ],
    [
        [
            ['Spain','Russia'],
            ['Croatia','Denmark']
        ],
    
        [
             ['Sweden','Switzerland'],
             ['Belgium','Japan']
        ]
    ]
]
```

```python
num_sims = 10000

#england finish 2nd
draw_1_sim = elo.sim_multiple_tournaments(tree=draw_1,
                                          n=num_sims,
                                          elo_dict=team_elos,)

#england finish 1st
draw_2_sim = elo.sim_multiple_tournaments(tree=draw_2,
                                          n=num_sims,
                                          elo_dict=team_elos,)

#england finish 1st - Colombia don't qualify
draw_3_sim = elo.sim_multiple_tournaments(tree=draw_3,
                                          n=num_sims,
                                          elo_dict=team_elos,)

#england finish 1st - Colombia don't qualify
draw_4_sim = elo.sim_multiple_tournaments(tree=draw_4,
                                          n=num_sims,
                                          elo_dict=team_elos,)
```

```python
for d, t in [(draw_1_sim,"England finish second in group"),
            (draw_2_sim,"England win group"),
            (draw_3_sim,"England win group (no Colombia)"),
            (draw_4_sim,"Germany still in"),]:
    elo.plot_counter(plt_data=d, title= t)
```

```python
print('England finish second in group: {:.1f}% win prob.'.format(draw_1_sim['England']))

print('England finish first in group: {:.1f}% win prob.'.format(draw_2_sim['England']))  

print('England finish first in group (no Colombia): {:.1f}% win prob.'.format(draw_3_sim['England'])) 

print('England finish first in group (no Colombia and Germany still in): {:.1f}% winn prob.'.format(draw_4_sim['England']))  
```
